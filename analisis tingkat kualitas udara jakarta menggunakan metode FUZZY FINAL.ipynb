```python
import numpy as np
import matplotlib.pyplot as plt
import skfuzzy as fuzz

# Range nilai PM2.5 dan PM10
pm25 = np.arange(0, 150, 1)
pm10 = np.arange(0, 350, 1)

# Fungsi Keanggotaan untuk PM2.5 dengan fuzz.trapmf
pm25_rendah = fuzz.trapmf(pm25, [0, 0, 5, 25])
pm25_sedang = fuzz.trapmf(pm25, [15, 30, 50, 70])
pm25_tinggi = fuzz.trapmf(pm25, [60, 90, 110, 150])

# Fungsi Keanggotaan untuk PM10 dengan fuzz.trapmf
pm10_rendah = fuzz.trapmf(pm10, [0, 0, 50, 100])
pm10_sedang = fuzz.trapmf(pm10, [80, 120, 150, 200])
pm10_tinggi = fuzz.trapmf(pm10, [180, 210, 250, 350])

# Fungsi Keanggotaan untuk ISPU
ISPU = np.arange(0, 150, 1)
ispu_baik = fuzz.trimf(ISPU, [0, 25, 50])
ispu_sedang = fuzz.trimf(ISPU, [50, 75, 100])
ispu_tidak_sehat = fuzz.trimf(ISPU, [100, 125, 150])

# Plot Fungsi Keanggotaan PM2.5
plt.figure(figsize=(8, 4))
plt.plot(pm25, pm25_rendah, 'g', linewidth=2, label='Rendah')
plt.plot(pm25, pm25_sedang, 'y', linewidth=2, label='Sedang')
plt.plot(pm25, pm25_tinggi, 'r', linewidth=2, label='Tinggi')
plt.xlabel('Nilai PM2.5 (g/m)')
plt.ylabel('Derajat Keanggotaan')
plt.title('Fungsi Keanggotaan PM2.5')
plt.legend()
plt.grid(True)
plt.ylim(0, 1.2)

plt.show()

# Plot Fungsi Keanggotaan PM10
plt.figure(figsize=(8, 4))
plt.plot(pm10, pm10_rendah, 'g', linewidth=2, label='Rendah')  
plt.plot(pm10, pm10_sedang, 'y', linewidth=2, label='Sedang')  
plt.plot(pm10, pm10_tinggi, 'r', linewidth=2, label='Tinggi')  
plt.xlabel('Nilai PM10 (g/m)')
plt.ylabel('Derajat Keanggotaan')
plt.title('Fungsi Keanggotaan PM10')
plt.legend()
plt.grid(True)
plt.ylim(0, 1.2)

plt.show()


# Plot Membership Functions
plt.figure(figsize=(8, 4))
plt.plot(ISPU, ispu_baik, 'g', linewidth=2, label='rendah')
plt.plot(ISPU, ispu_sedang, 'y', linewidth=2, label='Sedang')
plt.plot(ISPU, ispu_tidak_sehat, 'r', linewidth=2, label='tinggi')
plt.xlabel('Nilai ISPU')
plt.ylabel('Derajat Keanggotaan')
plt.title('Fungsi Keanggotaan Indeks Standar Pencemaran Udara (ISPU)')
plt.legend(loc='lower right')
plt.grid(True)
plt.ylim(0, 1.2)

plt.show()
```

```python
ISPU_lst = []

for i in range(len(df)):
    input_pm25 = df['pm25'][i]
    input_pm10 = df['pm10'][i]

    # Input: pm25 dan pm10 - DERAJAT KEANGGOTAAN
    pm25_Rendah = fuzz.interp_membership(pm25, pm25_rendah, input_pm25)
    pm25_Sedang = fuzz.interp_membership(pm25, pm25_sedang, input_pm25)
    pm25_Tinggi = fuzz.interp_membership(pm25, pm25_tinggi, input_pm25)

    pm10_Rendah = fuzz.interp_membership(pm10, pm10_rendah, input_pm10)
    pm10_Sedang = fuzz.interp_membership(pm10, pm10_sedang, input_pm10)
    pm10_Tinggi = fuzz.interp_membership(pm10, pm10_tinggi, input_pm10)

    # Aturan
    # Rule 1
    alpha_predikat1 = np.fmin(pm25_Rendah, pm10_Rendah)
    mf_rule1 = np.fmin(alpha_predikat1, ispu_baik)
    # Rule 2
    alpha_predikat2 = np.fmin(pm25_Rendah, pm10_Sedang)
    mf_rule2 = np.fmin(alpha_predikat2, ispu_sedang)
    # Rule 3
    alpha_predikat3 = np.fmin(pm25_Rendah, pm10_Tinggi)
    mf_rule3 = np.fmin(alpha_predikat3, ispu_tidak_sehat)
    # Rule 4
    alpha_predikat4 = np.fmin(pm25_Sedang, pm10_Rendah)
    mf_rule4 = np.fmin(alpha_predikat4, ispu_sedang)
    # Rule 5
    alpha_predikat5 = np.fmin(pm25_Sedang, pm10_Sedang)
    mf_rule5 = np.fmin(alpha_predikat5, ispu_tidak_sehat)
    # Rule 6
    alpha_predikat6 = np.fmin(pm25_Sedang, pm10_Tinggi)
    mf_rule6 = np.fmin(alpha_predikat6, ispu_tidak_sehat)
    # Rule 7
    alpha_predikat7 = np.fmin(pm25_Tinggi, pm10_Rendah)
    mf_rule7 = np.fmin(alpha_predikat7, ispu_tidak_sehat)
    # Rule 8
    alpha_predikat8 = np.fmin(pm25_Tinggi, pm10_Sedang)
    mf_rule8 = np.fmin(alpha_predikat8, ispu_tidak_sehat)
    
    aggregated = np.fmax(mf_rule1, np.fmax(mf_rule2, np.fmax(mf_rule3, np.fmax(mf_rule4, np.fmax(mf_rule5, np.fmax(mf_rule6, np.fmax(mf_rule7, mf_rule8)))))))

    ISPU_centroid = fuzz.defuzz(ISPU, aggregated, 'centroid')
    ISPU_lst.append(ISPU_centroid)
    print("ISPU (Centroid) - Data ke -", i, ':', ISPU_centroid)
```

```python
ISPU_lst = []

for i in range(len(df)):
    input_pm25 = df['pm25'][i]
    input_pm10 = df['pm10'][i]

    # Input: pm25 dan pm10 - DERAJAT KEANGGOTAAN
    pm25_Rendah = fuzz.interp_membership(pm25, pm25_rendah, input_pm25)
    pm25_Sedang = fuzz.interp_membership(pm25, pm25_sedang, input_pm25)
    pm25_Tinggi = fuzz.interp_membership(pm25, pm25_tinggi, input_pm25)

    pm10_Rendah = fuzz.interp_membership(pm10, pm10_rendah, input_pm10)
    pm10_Sedang = fuzz.interp_membership(pm10, pm10_sedang, input_pm10)
    pm10_Tinggi = fuzz.interp_membership(pm10, pm10_tinggi, input_pm10)

    # Aturan
    # Rule 1
    alpha_predikat1 = np.fmin(pm25_Rendah, pm10_Rendah)
    mf_rule1 = np.fmin(alpha_predikat1, ispu_baik)
    # Rule 2
    alpha_predikat2 = np.fmin(pm25_Rendah, pm10_Sedang)
    mf_rule2 = np.fmin(alpha_predikat2, ispu_sedang)
    # Rule 3
    alpha_predikat3 = np.fmin(pm25_Rendah, pm10_Tinggi)
    mf_rule3 = np.fmin(alpha_predikat3, ispu_tidak_sehat)
    # Rule 4
    alpha_predikat4 = np.fmin(pm25_Sedang, pm10_Rendah)
    mf_rule4 = np.fmin(alpha_predikat4, ispu_sedang)
    # Rule 5
    alpha_predikat5 = np.fmin(pm25_Sedang, pm10_Sedang)
    mf_rule5 = np.fmin(alpha_predikat5, ispu_tidak_sehat)
    # Rule 6
    alpha_predikat6 = np.fmin(pm25_Sedang, pm10_Tinggi)
    mf_rule6 = np.fmin(alpha_predikat6, ispu_tidak_sehat)
    # Rule 7
    alpha_predikat7 = np.fmin(pm25_Tinggi, pm10_Rendah)
    mf_rule7 = np.fmin(alpha_predikat7, ispu_tidak_sehat)
    # Rule 8
    alpha_predikat8 = np.fmin(pm25_Tinggi, pm10_Sedang)
    mf_rule8 = np.fmin(alpha_predikat8, ispu_tidak_sehat)
    
    aggregated = np.fmax(mf_rule1, np.fmax(mf_rule2, np.fmax(mf_rule3, np.fmax(mf_rule4, np.fmax(mf_rule5, np.fmax(mf_rule6, np.fmax(mf_rule7, mf_rule8)))))))

    # Cek apakah ada aturan yang memberikan kontribusi pada hasil agregasi
    if np.sum(aggregated) > 0:
        ISPU_centroid = fuzz.defuzz(ISPU, aggregated, 'centroid')
    else:
        # Jika tidak ada aturan yang memberikan kontribusi, gunakan metode centroid default
        ISPU_centroid = fuzz.centroid(ISPU, aggregated)
    
    ISPU_lst.append(ISPU_centroid)

    # Kategorisasi ISPU
    if ISPU_centroid <= 50:
        kategori_ispu = 'Rendah'
    elif 50 < ISPU_centroid <= 100:
        kategori_ispu = 'Sedang'
    else:
        kategori_ispu = 'tinggi'

    print("ISPU (Centroid) - Data ke -", i, ':', ISPU_centroid, '(', kategori_ispu, ')')
```
