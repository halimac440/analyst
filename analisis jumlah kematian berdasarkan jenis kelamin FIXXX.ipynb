```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('data.csv')
df
```

```python
df.info()
```

```python
df.groupby('kecamatan')['jumlah'].sum().plot(kind='bar')
```

```python
df.groupby('jenis_kelamin ')['jumlah'].sum().plot(kind='bar')
```

```python
df.groupby('kota_kabupaten')['jumlah'].sum().plot(kind='bar')
```

```python
x = df.groupby(['kota_kabupaten'])['jumlah'].sum().to_frame().rename(columns={'jenis_kelamin':'count'}).reset_index()
x
```

```python
abc = df.groupby(['kota_kabupaten','jenis_kelamin '])['jumlah'].sum().to_frame()
abc
```

```python
abc = abc.reset_index()
abc
```

```python
plt.figure(figsize=(10,7))
sns.barplot(x=abc['kota_kabupaten'], y=abc['jumlah'], hue=abc['jenis_kelamin '], ci=0, data=df)
```

```python
abd = df.groupby(['kecamatan','jenis_kelamin '])['jumlah'].sum().to_frame()
abd
```

```python
abd = abd.reset_index()
abd
```

```python
plt.figure(figsize=(10,7))
sns.barplot(x=abd['kecamatan'], y=abd['jumlah'], hue=abd['jenis_kelamin '], ci=0, data=df)
```
