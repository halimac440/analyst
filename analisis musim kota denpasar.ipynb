## buat kolom baru berisi kondisi cuaca yang telah di clustering

```python
a= df[['weather_description','weather_main','weather_id']]
a
```

```python
for i in df['weather_id'].unique():
    x= a[a['weather_id']==i]
    print(i,x['weather_description'].unique())
```

```python
x= []
for i in df['weather_id']:
    if i < 300:
        x.append('thunderstorm')
    elif i <400:
        x.append('drizzle')
    elif i<600:
        x.append('rain')
    elif i<800:
        x.append('fog')
    elif i<900:
        x.append('clear')
df['cuaca'] = x
df
```

```python
df['Year']= df.index.year
df['Month'] = df.index.month
df['Day']= df.index.day_name()

df
```

```python
pd.set_option('display.max_row',None)
a=df.groupby(['Year',"Month"])['weather_description'].agg(pd.Series.mode)
a
```

```python
a=df.groupby(['Year'])['weather_description'].agg(pd.Series.mode)
a=a.reset_index()
a
```

```python
a=pd.pivot_table(a, index='weather_description', values='weather_description', aggfunc='count')
a=a.reset_index()

a
```
